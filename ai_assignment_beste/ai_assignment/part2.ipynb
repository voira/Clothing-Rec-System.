```python
import numpy as np
import pandas as pd

df = pd.read_csv('sampled_data.csv')
df.head()
```

```python
# show unique values in each column except the barcode and image_url_1 column
for column in df.columns:
    if column != 'barcode' and column != 'image_url_1':
        print(column, df[column].unique())
```

```python
# define a scoring function for clothing reccomendation based on body type 
def score_clothing(item, body_type):
    score = 0
    if body_type == 'apple':
        if item["tops_fit"] in ["regular-fit-tops", 'oversized-tops']: 
            score += 4
        if item["neckline_type"] in ["v-neck", "sweetheart-neck", "round-v-neck", "shoulder-off"]:  
            score += 4
        if item["tops_length"] in ["thigh-length", "long"]:  
            score += 4
        if item["style"] in ["casuals", "feminine"]:
            score += 4
        if item["sleeve_type"] == "flare-sleeve":
            score += 4
        if item["waist_type"] == "mid-waist":
            score += 4    
        if item["bottoms_fit"] == "flare-bottoms":
            score += 4
        else:
            score += 1    
            
    elif body_type == 'hourglass':
        if item["tops_fit"] in ["regular-fit-tops", 'fitted-tops', 'bodycon-tops']: 
            score += 2
        if item["neckline_type"] in ["high-neck", "v-neck", "shoulder-off", 'square-neck']:  
            score += 1
        if item["tops_length"] in ["thigh-length", "short"]:
            score += 1
        if item["style"] in ["feminine", "sophisticated"]:
            score += 1
        if item["sleeve_type"] == "straight-sleeve":
            score += 2
        if item["waist_type"] == "high-waist":
            score += 2    
        else:
            score += 1

    elif body_type == 'rectangle':
        if item["tops_fit"] in ["regular-fit-tops", 'oversized-tops']:
            score += 2
        if item["neckline_type"] in ["v-neck", "sweetheart-neck", "round-v-neck", "shoulder-off"]:  
            score += 1
        if item["tops_length"] in ["thigh-length", "long"]: 
            score -= 1
        if item["style"] in ["casuals"]:
            score += 1
        if item["sleeve_type"] in ['puff-sleeve', "flare-sleeve", "flutter-sleeve"]:
            score += 1
        if item["more_attributes"] in ["belted", "buttons"]:
            score += 2    
        if item["bottoms_fit"] == "flare-bottoms":
            score -= 1
        else:
            score += 1 
    return score

# could be more detailed but I'm bored and I'm not getting paid to do this
# also adding the extra negatives would be valuable
```

```python
# change the dataframe into a dictionary list
df_dict = df.to_dict(orient='records')
df_dict[:5]
```

```python
from PIL import Image  
import requests
from io import BytesIO

# Rank clothing for an apple body type
body_type = "hourglass"

ranked_clothing = sorted(df_dict, key=lambda x: score_clothing(x, body_type), reverse=True)

# Print results
for i, item in enumerate(ranked_clothing):
    print(f"Rank {i+1}: {item}, Score: {score_clothing(item, body_type)}")
    # show image with the url of the top ranked clothing in the loop
    response = requests.get(ranked_clothing[i]['image_url_1'])
    img = Image.open(BytesIO(response.content))
    img.show()
    if i == 4:
        break

# just shirts are being reccomended at this point because they naturally have more attributes to score on
      
```
